### 2.1 Изучите распределение домов от наличия вида на набережную</br>
Постройте график</br>
Сделайте выводы

```python
data21 = df['waterfront'].apply(lambda x: 'Yes' if x else 'No').value_counts()

plt.pie(data21, autopct='%1.1f%%', labels=data21.index)
plt.title('Распределение домов от наличия вида на набережную');
```

#####  Большинство продаваемых домов без вида на набережную

### 2.2 Изучите распределение этажей домов

```python
data22 = df['floors'].value_counts().reset_index()
sns.barplot(x=data22['index'], y=data22['floors'])
plt.title('Распределение этажей домов');
plt.xlabel('Floors');
plt.ylabel('Counts');
```

##### Большинство продаваемых домов составляют одноэтажные. На втором месте двухэтажные

### 2.2 Изучите распределение состояния домов

```python
data23 = df['condition'].value_counts().reset_index()

sns.barplot(x=data23['index'], y=data23['condition'])
plt.title('Распределение состояния домов');
plt.xlabel('Condition');
plt.ylabel('Counts');
```

##### Большинство продаваемых домов среднего состояния. Далее смещение распределения идет  в сторону хорошего и очень хорошего сотстояния

## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.head()
```

```python
data31 = df.groupby('bedrooms').agg({'price': 'median'}).sort_values('price').round(2).reset_index()
sns.barplot(x=data31['bedrooms'], y=data31['price']);
plt.title('Стоимость домов от количества спален');
```

#####  Увеличение количества спален увеличивает стоимость дома

```python
data32 = df.groupby('waterfront').agg({'price': 'median'}).sort_values('price').round(2)
data32
plt.pie(data32.price, autopct='%1.1f%%', labels=data32.index)
plt.title('Стоимость домов от наличия вида на набережную');
```

#####  Наличие вида на набережную значительно увеличивает стоимость дома

```python
data33 = df.groupby('condition').agg({'price': 'median'}).sort_values('price').round(2).reset_index()
sns.barplot(x=data33['condition'], y=data33['price']);
plt.title('Стоимость домов от состояния дома');
```

##### Лучше состояние, больше стоимость

```python
data33 = df.groupby('yr_renovated').agg({'price': 'median'}).sort_values('price').round(2).reset_index()

plt.figure(figsize=(15,7))
sns.barplot(x=data33['yr_renovated'], y=data33['price']);
plt.title('Стоимость домов от состояния дома')
plt.xticks(rotation=90);
```

##### Дома с ремогтом дороже

```python
data33 = df.groupby('grade').agg({'price': 'median'}).sort_values('price').round(2).reset_index()

plt.figure(figsize=(15,7))
sns.barplot(x=data33['grade'], y=data33['price']);
plt.title('Стоимость домов от оценки уровня конструкции и дизайна');
```

##### Оценка уровня конструкции и дизайна значительно влияет на стоимость дома
