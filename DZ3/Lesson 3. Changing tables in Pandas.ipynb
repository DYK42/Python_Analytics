## Задача # 1
### 1.1</br>Скачать файл в уроке</br>Считать данные с помощью pandas</br>Вывести на экран первые 5 строк</br>

```python
df = pd.read_csv("kc_house_data.csv", encoding='windows-1251', parse_dates=['date'])
df.head()
```

### 1.2</br>Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.</br>Если реновации дома не было, то в новом признаке поставьте 0

```python
years = []

for i, row in df.iterrows():
    years.append((row['yr_renovated'] - row['yr_built']) if row['yr_renovated'] > 0 else 0)

df['delta_renovated'] = years
df.head()
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
from datetime import datetime

df['year_sale'] = df['date'].apply(lambda x: datetime.date(x).year)
df['month_sale'] = df['date'].apply(lambda x: datetime.date(x).month)
df.head()
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```

## Задача # 3
Составьте несколько сводных таблиц
### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален</br>Отсортируйте от меньшей стоимости к большей

```python
df31 = df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
df31
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df32 = df.groupby('condition').agg({'price': ['min', 'max', 'mean']})
df32
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df33 = df.pivot_table(index='waterfront',
                  columns='view', 
                  values='price',
                  aggfunc='count',
                  margins=True,
                  margins_name='Total').fillna(0)
df33
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df34 = df.pivot_table(index='floors',
                  columns='bedrooms', 
                  values='price',
                  aggfunc='count',
                  margins=True,
                  margins_name='Total').fillna(0)
df34
```

### Ответ: 1 этаж, 3 спальни

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df35 = df.pivot_table(index='condition',
                  columns='grade', 
                  values='price',
                  aggfunc='median').fillna(0)
df35
```
